```python
import re
import numpy as np
```

```python
forest = '()'
expression = r'\(\)'
replacement_split=  r'(()())'
replacement_end = r'(x)'

rand = np.random.rand()

matches = re.finditer(expression, forest)
last_end = 0
for match in matches: 
    start, end = match.span()
    print(start)
    if rand >= 0.5:
        forest = forest[last_end:start] + replacement_split + forest[start+1:]
        print(forest)
    else:
        forest = forest[last_end:start] + replacement_end+ forest[start+1:]
    
    
```
